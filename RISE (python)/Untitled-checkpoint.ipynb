```python
#12. Arrange string characters so lowercase letters come first
def arrange_string(s):
    lower = [ch for ch in s if ch.islower()]
    upper = [ch for ch in s if ch.isupper()]
    return "".join(lower + upper)

print("Arranged String:", arrange_string("PyNaT")) 
```

```python
#13.Count all letters, digits, and special symbols from a given string
str1 = "P@#yn26at^&t5on"
letters = sum(ch.isalpha() for ch in str1)
digits = sum(ch.isdigit() for ch in str1)
symbols = len(str1) - letters - digits

print("Letters:", letters)
print("Digits:", digits)
print("Symbols:", symbols)
```

```python
def f1():
    u=12
    def f2():
        nonlocal u
        u=u+2
        print(u)
    return f2
```

```python
print(f1())
```
